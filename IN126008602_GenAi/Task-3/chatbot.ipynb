**This is a chatbot, where I am using a Transformer Model which is hosted by some Inference provider (Groq for this model), as this was meant for purely testing without downloading the model locally**

```python
import os
from huggingface_hub import InferenceClient
from dotenv import load_dotenv
```

```python
load_dotenv()

HF_TOKEN = os.getenv("HF_TOKEN")
client = InferenceClient(api_key=HF_TOKEN)
```

```python
SYSTEM_PROMPT = {
    "role": "system",
    "content": (
        "You are a helpful AI assistant named JARVIS, created by Sayan."
        "Respond in simple, easy-to-understand language with a friendly tone. "
        "If you don't know something, say 'I don't know'. Do not make up answers."
        "The response should be in Plain text, don't use Markdown style"
    ),
}
```

```python
print("👋 Hello! I am your AI assistant. How can I help you today?")

while True:
    user_input = input("User: ")
    print(f"\nUser: {user_input}")

    if user_input.strip().lower() == "exit":
        print("Bot: Have a good day!")
        break

    full_message = [SYSTEM_PROMPT, {
        "role" : "user",
        "content" : user_input
    }]

    completion = client.chat.completions.create(
        model="openai/gpt-oss-20b:groq",
        messages=full_message,
    )

    reply = completion.choices[0].message.content

    print(f"Bot: {reply}")
```
