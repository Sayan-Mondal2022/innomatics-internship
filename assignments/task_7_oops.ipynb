## Problem 1: Smart Light Controller

**Problem Statement:**\
Create a class to control a smart light that can be turned ON or OFF and display its status.

**üåç Real-Time Use:**\
Smart home automation systems

**üì• Sample Input:**\
Light Name: Bedroom Light\
Action: ON

**üì§ Expected Output:**\
Bedroom Light is ON

```python
class Light:
    def __init__(self, light_name: str):
        self.light_name = light_name
        self.status = "OFF"

    # action() will change the status of the light to 'ON' or 'OFF'
    def action(self, act: str):
        if act.lower() == "on":
            self.status = "ON"
        else:
            self.status = "OFF"
        
```

```python
light_name = input("Enter the light name: ").strip()
action = input("Wht do you want to do? ('ON' or 'OFF') ")

if not light_name or not action:
    print("INVALID INPUT")
else:
    # CREATING the Object
    light = Light(light_name)
    light.action(action)

    print(f"Light name: {light_name} light")
    print(f"Action: {action}")

    print(f"\n{light.light_name} light is {light.status}")
```

## Problem 2: Employee ID Card System

**Problem Statement:**\
Create a class that stores employee name, ID, and department and displays ID card details.

**Real-Time Use:**\
Corporate HR systems

**üì• Sample Input:**\
Employee Name: Rahul\
Employee ID: EMP102\
Department: IT

**üì§ Expected Output:**\
Employee ID Card\
Name: Rahul\
ID: EMP102
Department: IT

```python
class Employee:
    def __init__(self, emp_name: str, emp_id: str, dept: str):
        self.emp_name = emp_name
        self.emp_id= emp_id
        self.dept = dept

    # Function to display the ID card
    def display_id(self):
        return f"{self.emp_name}\nID: {self.emp_id} Department: {self.dept}"
```

```python
name = input("Enter the employee name: ")
emp_id = input("Enter the employee id: ")
dept = input("Enter the employee department: ")

if not name or not emp_id or not dept:
    print("INVALID INPUT")
else:
    # CREATING the EMPLOYEE Object
    emp = Employee(name, emp_id, dept)

    print(f"Employee Name: {name}\nEmployee ID: {emp_id}\nDepartment: {dept}\n")

    print(emp.display_id())
```
