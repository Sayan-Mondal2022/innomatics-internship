## Problem 1: Smart Light Controller

**Problem Statement:**\
Create a class to control a smart light that can be turned ON or OFF and display its status.

**üåç Real-Time Use:**\
Smart home automation systems

**üì• Sample Input:**\
Light Name: Bedroom Light\
Action: ON

**üì§ Expected Output:**\
Bedroom Light is ON

```python
class Light:
    def __init__(self, light_name: str):
        self.light_name = light_name
        self.status = "OFF"

    # action() will change the status of the light to 'ON' or 'OFF'
    def action(self, act: str):
        if act.lower() == "on":
            self.status = "ON"
        else:
            self.status = "OFF"
        
```

```python
light_name = input("Enter the light name: ").strip()
action = input("Wht do you want to do? ('ON' or 'OFF') ")

if not light_name or not action:
    print("INVALID INPUT")
else:
    # CREATING the Object
    light = Light(light_name)
    light.action(action)

    print(f"Light name: {light_name} light")
    print(f"Action: {action}")

    print(f"\n{light.light_name} light is {light.status}")
```

## Problem 2: Employee ID Card System

**Problem Statement:**\
Create a class that stores employee name, ID, and department and displays ID card details.

**Real-Time Use:**\
Corporate HR systems

**üì• Sample Input:**\
Employee Name: Rahul\
Employee ID: EMP102\
Department: IT

**üì§ Expected Output:**\
Employee ID Card\
Name: Rahul\
ID: EMP102
Department: IT

```python
class Employee:
    def __init__(self, emp_name: str, emp_id: str, dept: str):
        self.emp_name = emp_name
        self.emp_id= emp_id
        self.dept = dept

    # Function to display the ID card
    def display_id(self):
        return f"{self.emp_name}\nID: {self.emp_id} Department: {self.dept}"
```

```python
name = input("Enter the employee name: ")
emp_id = input("Enter the employee id: ")
dept = input("Enter the employee department: ")

if not name or not emp_id or not dept:
    print("INVALID INPUT")
else:
    # CREATING the EMPLOYEE Object
    emp = Employee(name, emp_id, dept)

    print(f"Employee Name: {name}\nEmployee ID: {emp_id}\nDepartment: {dept}\n")

    print(emp.display_id())
```

## Problem 3: Mobile Contact Record

**Problem Statement:**\
Create a class to store contact name and phone number and display contact information.

**Real-Time Use:**\
Mobile contact applications

**üì• Sample Input:**\
Contact Name: Anita\
Phone Number: 9876543210

**üì§ Expected Output:**\
Contact Saved\
Name: Anita\
Phone: 9876543210

```python
class ContactDetail:
    def __init__(self, name: str, phone_no: str):
        self.name = name
        self.phone_no = phone_no

    def display_info(self):
        return f"Name: {self.name}\nNumber: {self.phone_no}"
```

```python
name = input("Enter the contact name: ").strip()
phone_no = input("Enter the phone number: ").strip()

if not name or not phone_no:
    print("INVALID INPUT")
else:
    print(f"Contact name: {name}\nPhone number: {phone_no}")

    contact_details = ContactDetail(name, phone_no)
    print("\nContact Saved")

    print(contact_details.display_info())
```

## Problem 4: Product Price Tag Generator

**Problem Statement:**\

**üì§ Expected Output:**\

```python
class Product:
    def __init__(self, product_name: str, price: float):
        self.product_name = product_name
        self.price = price

    def show_price(self):
        return f"Product: {self.product_name}\nPrice: ‚Çπ{self.price}"
```

## Problem 5: Movie Rating Display System

**Problem Statement:**\
Create a class that stores movie name and rating and displays movie details.

**Real-Time Use:**\
Streaming platforms

**üì• Sample Input:**\
Movie Name: Inception\
Rating: 4.8

**üì§ Expected Output:**\
Movie: Inception\
Rating: 4.8 / 5

```python
class Movie:
    def __init__(self, movie_name: str, rating: float):
        self.movie_name = movie_name
        self.rating = min(rating, 5.0)

    def display_movie(self):
        return f"Movie: {self.movie_name}\nRating: {self.rating} / 5"
```

```python
movie_name = input("Enter the movie name: ").strip()
rating = float(input("Enter the movie rating: ").strip())

if not movie_name or not rating:
    print("INVALID INPUT")
else:
    print(f"Movie Name: {movie_name}\nRating: {rating}\n")

    movie = Movie(movie_name, rating)

    print(movie.display_movie())
```

## Problem 6: Delivery Address Manager

**Problem Statement:**\
Create a class that stores customer name and delivery address and prints delivery details.

**Real-Time Use:**\
Courier and logistics platforms

**üì• Sample Input:**\
Customer Name: Suman\
Address: Hyderabad

**üì§ Expected Output:**\
Delivery Details\
Customer: Suman\
Address: Hyderabad

```python
class DeliveryAddress:
    def __init__(self, customer_name: str, address: str):
        self.customer_name = customer_name
        self.address = address

    def delivery_details(self):
        return f"Delivery Details\nCustomer: {self.customer_name}\nAddress: {self.address}"
```

```python
customer_name = input("Enter the customer name: ").strip().capitalize()
address = input("Enter the address: ").strip()

if not customer_name or not address:
    print("INVALID INPUT")
else:
    print(f"Customer Name: {customer_name}\nAddress: {address}\n")

    delivery_address = DeliveryAddress(customer_name, address)

    print(delivery_address.delivery_details())
```
