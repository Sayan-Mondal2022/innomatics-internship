### Q.1 User Login Check

**Problem Statement:**\
Given a username and password, check whether login is successful.

Login system needs the following inputs
1. Username - This is a string (can contain alphanumeric values)
2. Password - This is a string (can contain alphanumeric values)

**Requirements:** \
Print *"Login Successful"* if both username and password match \
Otherwise print *"Invalid Credentials"*

**Real-World Application:** Authentication systems

```python
# Using a function, in-order to use the same function everytime
def valid_credentials(username: str, password: str) -> str:
    # Credentials stored:
    USERNAME = "admin"
    PASSWORD = "1234"

    # inputs for the login system:
    print(f"Username is: {username}")
    print(f"User password is: {password}")

    # Performing the check using the conditional statements
    if username == USERNAME and password == PASSWORD:
        return "Login Successful"
    else:
        return "Invalid Credentials"
```

```python
# Invalid credentials
print(valid_credentials("abc", "5691"))

print("\n")

# Valid credentials
print(valid_credentials("admin", "1234"))
```

### Q.2  Pass / Fail Analyzer

**Problem Statement:** \
Analyze student results.

**Requirements:**
- A student passes if marks â‰¥ 50
- Count the total number of pass students
- Count the total number of fail students
- Print the final result clearly

**Real-World Application:** Academic evaluation systems

```python
# Marks for 5 students
marks = [45, 78, 90, 33, 60]

# This is to keep the count ot pass and failed students
total_passed, total_failed = 0, 0

# Using loop to iter through all the marks
for mark in marks:
    if mark >= 50:
        total_passed += 1
    else:
        total_failed += 1

# This is to get the total number of students
n = len(marks)

print("Academic result:\n")
print(f"Total number of students: {n}")
print(f"Total number of students passed: {total_passed}")
print(f"Total number of students failed: {total_failed}")
print(f"Passing percentage: {(total_passed / n) * 100 :.2f}%")
print(f"Fail percentage: {(total_failed / n) * 100 :.2f}%")
```

### Q.3 Simple Data Cleaner 

**Problem Statement:**\
Clean and standardize user names.

**Requirements:**
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list

**Real-World Application:** Data preprocessing before analysis


```python
# Raw Data:
names = [" Alice ", "bob", " CHARLIE "]

# Will be used to store the Cleaned Names
cleaned_names = []

for name in names:
    # strip() -> this is used to removed the whitespaces from both left and right
    # lower() -> this is used to change every character to smaller
    name = name.strip().lower()

    # storing the cleaned names into my new list
    cleaned_names.append(name)

print(f"Cleaned names:\n{cleaned_names}")
```

### Q.4 Message Length Analyzer

**Problem Statement:**\
Analyze message sizes.

**Requirements:**
- Print the length of each message
- Flag messages longer than 10 characters

**Real-World Application:** Text filtering and validation systems

```python
# Messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Will be using sets to perform text filtration
flagged_msgs = set()
normal_msg = set()

# Loop to iter through all the messages
for msg in messages:
    # Conditional statement to Flag longer messages
    if len(msg) > 10:
        flagged_msgs.add(msg)
    else:
        normal_msg.add(msg)

print("Normal msgs:")
for msg in normal_msg:
    print(f"{msg} -> length {len(msg)}")

print("\nFlagged messages:")
for msg in flagged_msgs:
    print(f"{msg} -> length {len(msg)}")
```

### Q.5 Error Message Detector

**Problem Statement:**\
Detect error messages from system logs.

**Requirements:**
- Count the number of "ERROR" entries
- Print the total error count

**Real-World Application:** Monitoring and log analysis systems


```python
# Log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Will be using dictionary to maintain the count
freq = {}

for log in logs:
    # Using get method, which is in-built for dictionaries in python
    freq[log] = 1 + freq.get(log, 0)

print(f"The total number of 'ERROR' is: {freq['ERROR']}")
```
