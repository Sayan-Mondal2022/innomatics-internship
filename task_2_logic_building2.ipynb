### Problem 1: Unique Words in a Sentence

**Problem Statement:**

Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves

**Real-Time Applications**
1. Text analytics
2. Natural Language Processing (NLP) preprocessing
3. Search engine indexing
4. Chat analysis systems

```python
# Function to get all the unique words in a sentence:
def getUniqueWords(words: list[str]) -> set[str]:
    # Using a SET DATA STRUCTURE.
    uniqueWords = set()

    for word in words:
        uniqueWords.add(word.lower())

    return uniqueWords
```

```python
# Getting the input from the user.
sentence = input("Enter a sentence: ")
uniqueWords = getUniqueWords(sentence.split())

print(f"Sentence: {sentence}\nUnique Words: {uniqueWords}")
print(f"Every word count: {len(sentence.split())}\nUnique words count: {len(uniqueWords)}")
```

### Problem 2: Highest Salary from Employee Data

**Problem Statement**\
Store employee names and their salaries, and determine which employee has the highest salary.

**Input Format**\
A dictionary where:
1. Keys represent employee names
2. Values represent employee salaries

**Real-Time Applications**
- HR management systems
- Payroll analysis



```python
# Defining a Function to take n-Employee inputs.

def takeEmployeeEntry(n: int) -> dict:
    # Defining employees as a Empty dict
    employees = {}      # name -> salary

    # Getting a inout for n employees:
    for i in range(n):
        name = input("Enter the employee name: ")
        salary = float(input("Enter the salary: "))

        print(f"{i + 1}. Employee Name: {name}, Salary: {salary}")

        # Storing the values in Employees dict.
        employees[name] = salary

    return employees
```

```python
# Once the Data is collected from the user: 
def getHighestSalary(employees: dict):
    # This will be used to store the Maximum Salary
    maxSalary = 0
    name = ""

    for emp in employees:
      if maxSalary < employees[emp]:
        maxSalary = employees[emp]
        name = emp
    
    return [name, maxSalary]
```

```python
# Asking user to Enter number of Employees:
n = int(input("Enter the number of Employees: "))
print(f"Number of Employee are: {n}\n")

employees = takeEmployeeEntry(n)
name, maxSalary = getHighestSalary(employees)

print(f"\nEmployee name: {name}\nHighest Salary is: {maxSalary}")
```

### Problem 3: Find Maximum and Minimum Values

**Problem Statement:**\
Find the maximum and minimum values from a list of numbers.

**Input Format:**\
A list of integers where each element represents a numeric value.
Example Input
numbers = [45, 22, 89, 10, 66]

**Expected Output:**\
List: [45, 22, 89, 10, 66]  
Max: 89  
Min: 10

**Real-Time Applications:**
- Sales data analysis
- Temperature monitoring systems
- Performance metric evaluation

```python
# Getting list of integers from user:
nums = list(map(int, input().split()))

# Initially assigning the first value from the list
maxNum, minNum = nums[0], nums[0]

# Using FOR LOOP to iter:
for n in nums[1:]:
    maxNum = max(maxNum, n)
    minNum = min(minNum, n)

print(f"List: {nums}\nMax: {maxNum}\nMin: {minNum}")
```

### Problem 4: Count Products Above a Price Threshold

**Problem Statement:**\
Count how many products have a price greater than 1000.


**Input Format:**\
A list of integers where each element represents the price of a product.

**Example Input:**\
prices = [450, 1200, 899, 1500, 300]

**Expected Output:**\
Products above 1000: 2

**Real-Time Applications:**
- E-commerce product filtering
- Budget analysis


```python
# Getting the user Input:
prices = list(map(int, input().split()))
print(f"Prices: {prices}")

result = 0

for price in prices:
    if price > 1000:
        result += 1

print(f"Products above 1000: {result}")
```

### Problem 5: Calculate Attendance Percentage

**Problem Statement:**\
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

**Example Input:**\
attendance = ["P", "P", "A", "P", "P"]

**Expected Output:**\
Attendance Percentage: 80.0

**Real-Time Applications:**
- Internship tracking systems
- HR attendance management systems

```python
# Getting the user input:
attendance = input().split()
print(f"Attendance: {attendance}")

present = 0
for a in attendance:
    if a.upper() == "P":
        present += 1

print(f"Attendance Percentage is: {(present / len(attendance)) * 100:.1f}")
```

## Problem 6: Remove Duplicate Phone Numbers

**Problem Statement:**\
You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

**Input Format:**\
A list of integers where each integer represents a phone number.

**Example Input:**\
phone_numbers = [9876543210, 9123456789, 9876543210]

**Expected Output:**\
Unique phone numbers: {9876543210, 9123456789}

**Real-Time Applications:**
- Contact management applications
- Customer Relationship Management (CRM) systems
- Marketing databases


```python
# Get the input from the user.
phoneNumbers = input().split()
print(f"Phone numbers: {phoneNumbers}")

# Using set to maintain unique numbers
uniquePhoneNumbers = set()

for number in phoneNumbers:
    uniquePhoneNumbers.add(number)

print(f"Unique phone numbers are: {uniquePhoneNumbers}")
```

### Problem 7: Count Character Frequency

**Problem Statement:**\
Count how many times each character appears in a given string.

**Expected Output:**\
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}

**Real-Time Applications:**
- Text analysis
- Password strength validation


```python
word = input("Enter a word: ")
print(f"Word: {word}")

# Using a Dictionary to keep a frequency count.
freq = {}

for char in word:
    # In-case if the user gives capital letters as well
    char = char.lower()
    
    freq[char] = 1 + freq.get(char, 0)

print(f"Frequency count: {freq}")
```

### Problem 9: Check if a Key Exists in a Dictionary

**Problem Statement:**\
Check whether a specific key exists in a dictionary.

**Expected Output:**\
Employee exists

**Real-Time Applications:**
- User validation systems
- Configuration verification


```python
# First adding the Employees and their Salaries:
employees = {}

# Getting a inout for 5 employees:
for i in range(5):
    name = input("Enter the employee name: ").lower()
    salary = float(input("Enter the salary: "))

    print(f"{i + 1}. Employee Name: {name}, Salary: {salary}")

    # Storing the values in Employees dict.
    employees[name] = salary

print(f"Employees are: {employees}")
```

```python
# Now taking the user input to find whether the Employee exist in the Employee Dictionary or not
employeeNames = input("Enter the Employee names: ").split()
print(f"Entered employee names are: {employeeNames}\n")

for empName in employeeNames:
    empName = empName.lower()
    
    if empName in employees:
        print(f"Employee: {empName} exist in the company")
    else:
        print(f"Employee: {empName} doesn't exist in the company")
```

### Problem 10: Calculate Average Marks

**Problem Statement:**\
Calculate the average marks of a student from a list of marks.

**Real-Time Applications:**
- Student performance evaluation
- Result processing systems


```python
# Input marks from the user:
marks = list(map(int, input().split()))

total = 0
n = len(marks)

for m in marks:
    total += m

print(f"Marks: {marks}\nAverage marks: {(total / n):.2f}")
```
